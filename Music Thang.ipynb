```python
import numpy
import urllib ## should use urllib.request for urlopen
import ast
import json
import os
import eyed3
import sys
import re
```

```python
artistName = input("Enter artist name: ")

artistName = artistName.replace(" ", "+")
print("Name is: ", artistName)
```

```python
#data = [d for d in data if 'user/ageInSeconds' in d and d['user/ageInSeconds'] < 60*60*24*365*100]
ituneURL = urllib.request.urlopen("https://itunes.apple.com/search?term="+artistName+"&limit=25")
```

```python
d = ituneURL.read()
```

```python
d
```

```python
encoding = ituneURL.info().get_content_charset('utf-8')
json_obj = json.loads(d.decode(encoding))
```

```python
json_obj
```

```python

for i in range(len(json_obj["results"])):
    print(json_obj["results"][i]["trackName"])
    

print(len(json_obj["results"]))
```
